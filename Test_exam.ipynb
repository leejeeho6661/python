```python
import pymysql
conn = pymysql.connect(host='127.0.0.1', user='root',password='',port=13306,database='mydb',charset='utf8')
cursor = conn.cursor(pymysql.cursors.DictCursor)
def selectlist():
    selectSql = '''select id,pwd,name,req_date from member'''
    cursor.execute(selectSql)
    
    return cursor.fetchall()
def displayMember():
    rows = selectlist()
    if len(rows) == 0:
          print('회원이 존재하지 않습니다.')
    else:
        for row in rows:
            print("{}\t{}\t{}\t{}".format(row['id'],row['pwd'],row['name'],row['req_date']))
    
def memberSearch(id):
    selectSql = '''select id,pwd,name,req_date from member where id=%s'''
    cursor.execute(selectSql,(id))
    row = cursor.fetchone()
    print(type(row))
    if row == None:
        print('일치하는 회원이 존재하지 않습니다.')
    else:
        print(id+"회원이 존재합니다.")
def insertMember():
    id = input('등록할 아이드를 입력하세요 -> ')
    pwd = input('비밀번호를 입력하세요 -> ')
    name = input('이름을 입력하세요 -> ')
    sql = '''insert into member(id,pwd,name,req_date) values(%s,%s,%s,now())'''
    try:
        cursor.execute(sql,(id,pwd,name))
        conn.commit()
        print('회원등록이 완료되었습니다.')
    except:
        print('등록에 실패하였습니다.')
        
def deleteMember():
    id = input('삭제할 아이디를 입력하세요')
    sql = '''delete from member where id=%s'''
    try:
        cursor.execute(sql,(id))
        conn.commit()
        print('삭제 완료되었습니다.')
    except:
        print('삭제에 실패하였습니다.')
def endSql():
    cursor.close()
    conn.close()
def main():
    print('#'*30)
    print('# 회원 관리 프로그램')
    print('#'*30)
    
    while True:
        try:
            print('# 1. 전체 회원 목록 보기')
            print('# 2. 회원 검색')
            print('# 3. 회원 등록')
            print('# 4. 회원 삭제')
            print('# 5. 종료')
            cmd = int(input('메뉴를 선택하세요(1~5) -> ')) 
        except ValueError:
            print('1~5의 값을 입력해주세요')
            continue
            
        if cmd ==1:
            print('회원목록입니다.')
            displayMember()
            continue
        elif cmd ==2:
            print('회원검색입니다.')
            memberId = input('검색할 회원을 입력하세요 ->')
            memberSearch(memberId)
            continue
        elif cmd ==3:
            print('회원등록입니다.')
            insertMember()
            continue
        elif cmd ==4:
            print('회원삭제입니다.')
            deleteMember()
            continue
        elif cmd ==5:
            endSql()
            break
```

```python
main()
```
